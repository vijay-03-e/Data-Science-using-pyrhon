# Interview Questions:

1. What is the difference between precision and recall?

Precision measures the proportion of positive predictions that are actually true positives.

Recall measures the proportion of actual positives that are correctly predicted as positive.

In other words, precision is about the accuracy of the positive predictions, while recall is about the completeness of the positive predictions.

2. What is cross-validation, and why is it important in binary classification?

Cross-validation is a technique used to evaluate the performance of a machine learning model on unseen data. 
It is important in binary classification because it helps to prevent overfitting and provides a more accurate estimate of the model's generalization performance.
