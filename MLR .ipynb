##1.What is Normalization & Standardization and how is it helpful?

Normalization & Standardization are two common techniques used in data preprocessing to transform data into a more suitable format for machine learning algorithms.

Normalization scales the values of each feature to a range between 0 and 1. This is useful when the features have different units or scales, as it ensures that no single feature dominates the model.

Standardization scales the values of each feature to have a mean of 0 and a standard deviation of 1. This is useful when the features are normally distributed, as it helps to improve the performance of many machine learning algorithms.

Both normalization and standardization can be helpful in improving the performance of machine learning models by:

* Reducing the influence of outliers
* Making the data more comparable
* Improving the convergence of gradient-based algorithms

##2.What techniques can be used to address multicollinearity in multiple linear regression?

1.Feature selection:
  - Identify and remove features that are highly correlated with each other.
  - This can be done using correlation analysis or by using a technique such as principal component analysis (PCA).

2.Ridge regression:
  - This is a type of linear regression that adds a penalty term to the cost function to discourage large coefficients.
  - This can help to reduce the impact of multicollinearity on the model.

3.Lasso regression:
  - This is another type of linear regression that adds a penalty term to the cost function to encourage the coefficients to be sparse.
  - This can help to reduce the impact of multicollinearity on the model and can also lead to feature selection.
